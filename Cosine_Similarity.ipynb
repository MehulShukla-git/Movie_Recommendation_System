```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
# ==============================
# COSINE SIMILARITY CLASSIFICATION (80:20 & 70:30)
# ==============================

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, f1_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# --------------------------------
# STEP 1 : Create Classification Label
# --------------------------------

df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
df['label'] = df['vote_average'].apply(lambda x: 1 if x >= 7 else 0)

# --------------------------------
# STEP 2 : Prepare Data
# --------------------------------

X = tfidf_matrix
y = df['label']

# --------------------------------
# FUNCTION TO EVALUATE COSINE MODEL
# --------------------------------

def evaluate_cosine(test_size):

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    y_pred = []
    y_prob = []

    for i in range(X_test.shape[0]):

        sim = cosine_similarity(X_test[i], X_train).flatten()

        idx = sim.argmax()

        y_pred.append(y_train.iloc[idx])

        # Use similarity score as probability (important fix)
        y_prob.append(sim[idx])

    y_pred = np.array(y_pred)
    y_prob = np.array(y_prob)

    accuracy = accuracy_score(y_test, y_pred) * 100
    precision = precision_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)

    return accuracy, precision, f1, roc_auc, y_test, y_prob


# --------------------------------
# STEP 3 : Evaluate for 80:20 Split
# --------------------------------

acc1, pre1, f11, roc1, y_test1, y_prob1 = evaluate_cosine(0.2)

# --------------------------------
# STEP 4 : Evaluate for 70:30 Split
# --------------------------------

acc2, pre2, f12, roc2, y_test2, y_prob2 = evaluate_cosine(0.3)

# --------------------------------
# STEP 5 : Create Result Tables
# --------------------------------

table_80_20 = pd.DataFrame({
    "Metric": ["Accuracy (%)", "Precision", "F1 Score", "ROC AUC"],
    "Value": [acc1, pre1, f11, roc1]
})

table_70_30 = pd.DataFrame({
    "Metric": ["Accuracy (%)", "Precision", "F1 Score", "ROC AUC"],
    "Value": [acc2, pre2, f12, roc2]
})

# --------------------------------
# STEP 6 : Print Tables
# --------------------------------

print("\n========== COSINE SIMILARITY 80:20 RESULTS ==========\n")
print(table_80_20)

print("\n========== COSINE SIMILARITY 70:30 RESULTS ==========\n")
print(table_70_30)

# --------------------------------
# STEP 7 : ROC CURVES (Both Splits)
# --------------------------------

fpr1, tpr1, _ = roc_curve(y_test1, y_prob1)
fpr2, tpr2, _ = roc_curve(y_test2, y_prob2)

plt.figure(figsize=(6,4))
plt.plot(fpr1, tpr1, label="80:20 Split")
plt.plot(fpr2, tpr2, label="70:30 Split")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Cosine Similarity")
plt.legend()
plt.show()

# --------------------------------
# STEP 8 : Movie Recommendation Function
# --------------------------------

def recommend_movies():

    movie = input("\nEnter movie name: ")

    if movie not in indices:
        print("Movie not found")
        return

    idx = indices[movie]

    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()

    similar_idx = sim_scores.argsort()[::-1][1:11]

    recommendations = df['title'].iloc[similar_idx]

    print("\nRecommended Movies:\n")

    for m in recommendations:
        print(m)

# --------------------------------
# STEP 9 : Run Recommendation System
# --------------------------------

recommend_movies()
```
